```python
def createVocabList(dataSet):
    vocabSet = set([])
    for document in dataSet:
        vocabSet = vocabSet | set(document) # set() : remove duplicated word  '|'는 없는 것만 합해준다!!!
        # ex) ('dog','flea','has','help','my','please','problem') sort도 된다. 이걸 반복해 추가해 간다!!  
    return list(vocabSet)
```

```python
def setOfWords2Vec(vocabList, inputSet): # 단어의 벡터화!! 
    returnVec = [0]*len(vocabList)
    for word in inputSet:
        if word in vocabList:
            returnVec[vocabList.index(word)] = 1
        else: print "the word: %s is not in my Vocabulary!" % word
    return returnVec
```

### Listing 4.2 Naïve Bayes classifier training function

```python
def trainNB0(trainMatrix,trainCategory):
    numTrainDocs = len(trainMatrix)
    numWords = len(trainMatrix[0])
    pAbusive = sum(trainCategory)/float(numTrainDocs)
    # 확률 초기화
    p0Num = zeros(numWords); p1Num = zeros(numWords)
    p0Denom = 0.0; p1Denom = 0.0
    
    for i in range(numTrainDocs):
        if trainCategory[i] == 1:
            # 벡터 추가
            p1Num += trainMatrix[i]
            p1Denom += sum(trainMatrix[i])
        else:
            p0Num += trainMatrix[i]
            p0Denom += sum(trainMatrix[i])
    # 원소 나누기
    p1Vect = p1Num/p1Denom #change to log() # P(Wi/C1)
    p0Vect = p0Num/p0Denom #change to log() # P(Wi/C0)
    return p0Vect,p1Vect,pAbusive
```

```python

def classifyNB_lec(vec2Classify, p0Vec, p1Vec, pClass1): 
    p1 = sum(vec2Classify * p1Vec) + log(pClass1)    #element-wise mult
    p0 = sum(vec2Classify * p0Vec) + log(1.0 - pClass1)
    if p1 > p0:
        return 1
    else: 
        return 0
```

- vec2Classify, P(c1|w[t]) 에서 w[t]를 나타낸다. 테스트할 word들의 vector!!
    - 이 테스트할 word 들이 테스트 하고 싶은 글에 들어 있는 거지!! 
    - 그래서 그 글을 테스트 하는 거다!! 


- pClass1, P(c1) 이다! - 폭력적인 글일 확률


- P(c1|w[t]) = (P(w|c1) * P(c1)) / P(w)      
    - P(w) = P(w1,w2 ... ,w31)


- sum(vec2Classify * p1Vec) + log(pClass1)
    - 덧셈을 하는 이유는 log를 취해서 p1Vec을 구했었기 때문에!!     


- Q. 그러면 32개의 정해진 단어에 대해서만 확률을 가지고 있기 때문에 테스트 하는 문장들도 그 32개의 단어만 count 해서 TEST 하겠군! ㅇㅇ
